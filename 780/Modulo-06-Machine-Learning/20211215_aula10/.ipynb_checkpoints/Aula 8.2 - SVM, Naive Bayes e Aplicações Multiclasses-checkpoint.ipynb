## Exercícios

**1)** Utilizando do modelo SVM, faça a classificação do dataset _Insurance.csv_ para fazer as predições!

<br>

_Dica:_ Esse é um caso multiclasse, então utilize o __One vs One__ e o __One vs Rest__ para obter o resultado e compare se algum dos dois performa melhor.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
insurance = pd.read_csv('../datasets/insurance.csv')
```

```python
insurance.info()
```

```python
sns.jointplot(data=insurance, x='age', y='charges')
```

```python
insurance['target'] = np.where(insurance['charges'] < 10000, 0, np.where(insurance['charges'] < 30000, 1, 2))
```

```python
insurance['target'].value_counts()
```

```python
insurance['flag_sex'] = np.where(insurance['sex'] == 'female', 1, 0)
```

```python
insurance['is_smoker'] = np.where(insurance['smoker'] == 'yes', 1, 0)
```

```python
X = insurance.drop(['charges', 'target', 'sex', 'smoker'], axis = 1)
y = insurance["target"]
```

```python
X.head()
```

```python
X_with_dummies = pd.get_dummies(X, prefix_sep='_', columns=['region'])
```

```python
X_with_dummies.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_with_dummies, 
                                                    y, 
                                                    test_size=0.5, 
                                                    random_state=42
                                                   # ,stratify = y
                                                   )
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
```

```python
X_test_std = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
```

```python
model = OneVsOneClassifier(SVC(random_state=42)).fit(X_train_std, y_train)
```

```python
y_pred = model.predict(X_test_std)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

plot_confusion_matrix(conf_mat=cm)
plt.show()
```

Testando agora o __One vs Rest__:

```python
model = OneVsRestClassifier(SVC(random_state=42)).fit(X_train_std, y_train)

y_pred = model.predict(X_test_std)

print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

plot_confusion_matrix(conf_mat=cm)
plt.show()
```
