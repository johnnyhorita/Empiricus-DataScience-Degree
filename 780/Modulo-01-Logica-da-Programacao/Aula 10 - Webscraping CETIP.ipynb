### Juntando tudo em uma única função

```python
from selenium import webdriver
import bs4
import time
import datetime as dt

def modifica_data(driver, data, inicial=True):    
    dia_elemento = 'DT_DIA_DE' if inicial == True else 'DT_DIA_ATE'
    dia = driver.find_element_by_name(dia_elemento)
    dia.clear()
    time.sleep(1)
    dia.send_keys(data.day)
    
    mes_elemento = 'DT_MES_DE' if inicial == True else 'DT_MES_ATE'
    mes = driver.find_element_by_name(mes_elemento)
    mes.clear()
    time.sleep(1)  # Não precisa do sleep
    mes.send_keys(data.month)
    
    ano_elemento = 'DT_ANO_DE' if inicial == True else 'DT_ANO_ATE'
    ano = driver.find_element_by_name(ano_elemento)
    ano.clear()
    time.sleep(1)  # Não precisa do sleep
    ano.send_keys(data.year)
    
def click_checkbox(driver, itens):
    dicionario_item_para_checkbox = {
        "media": 'chk_M1',
        "minima": 'chk_M2',
        "maxima": 'chk_M3',
        "moda": 'chk_M4',
        "desvio_padrao": 'chk_M5',
        "taxa_selic": 'chk_M6'
    }
    for item in itens:
        print(f'Selecionando opção {item}')
        nome_checkbox = dicionario_item_para_checkbox[item]
        driver.find_element_by_name(nome_checkbox).click()
        
def pesquisa(driver):
    pesquisar_xpath = '/html/body/div/div/div/div/div/div/form/div[6]/div/a[2]'
    driver.find_element_by_xpath(pesquisar_xpath).click()
    
def muda_pagina(driver):
    botao = driver.find_elements_by_class_name('btProxAnt')[-1]
    texto_botao = botao.get_attribute('text').strip()
    if 'Próximo' == texto_botao:
        botao.click()
    else:
        raise Exception('Não há novas páginas')
        
def extrai_dados(elementos, cabecalho):
    respostas = []
    linha = []
    classe_anterior = None
    for i, elemento in enumerate(elementos):
        classe = elemento['class'][0]
        texto = elemento.text.strip()
        if classe.startswith('Cabecalho'):
            pass
        else:
            if classe != classe_anterior and linha:
                if linha != cabecalho:
                    respostas.append(linha)
                if not cabecalho:
                    cabecalho = linha

                linha = []
                linha.append(texto)
            else:
                linha.append(texto)
        classe_anterior = classe
    return respostas, cabecalho

def connecta_cetip():
    driver = webdriver.Chrome('./chromedriver')
    cetip_url = 'http://estatisticas.cetip.com.br/astec/series_v05/paginas/lum_web_v05_template_informacoes_di.asp?str_Modulo=completo&int_Idioma=1&int_Titulo=6&int_NivelBD=2'
    driver.get(cetip_url)
    return driver

def raspagem_cetip(inicio='1986-03-06',
                  fim=None,
                  itens=['media',
                         'minima',
                         'maxima',
                         'moda',
                         'desvio_padrao',
                         'taxa_selic']
                  ):
    inicio = dt.datetime.strptime(inicio, '%Y-%m-%d')
    fim = dt.datetime.strptime(fim, '%Y-%m-%d') if fim else dt.date.today()
    
    print('Conectando ao CETIP')
    print('-'*32)
    driver = connecta_cetip()
    print(f'Modificando data de inicio {inicio:%Y-%m-%d}')
    print('-'*32)
    modifica_data(driver, inicio)
    print(f'Modificando data de fim {fim:%Y-%m-%d}')
    print('-'*32)
    modifica_data(driver, fim, inicial=False)
    
    click_checkbox(driver, itens)
    print('-'*32)
    print('Clicando em pesquisar')
    print('-'*32)
    pesquisa(driver)
    respostas = []
    cabecalho = []
    numero = 0
    while True:
        try:
            numero += 1
            element = driver.find_element_by_tag_name('body')
            resposta = element.get_attribute('innerHTML')
            site = bs4.BeautifulSoup(resposta, 'html.parser')
            elementos = site.find_all('td')
            print(f'Extraindo dados da pagina {numero}')
            print('-'*32)
            resposta, cabecalho = extrai_dados(elementos, cabecalho)
            print(len(resposta))
            respostas.extend(resposta)
            print('Movendo de pagina')
            print('-'*32)
            time.sleep(3)
            muda_pagina(driver)
        except:
            print('Fim')
            break
    return respostas
```
