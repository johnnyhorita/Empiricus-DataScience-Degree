```python
import yfinance as yf
# Ignore esse import
import matplotlib.pyplot as plt

acoes = ['EMBR3.SA', 'AZUL4.SA', 'CSNA3.SA', 'MGLU3.SA']
for acao in acoes:
    yf_acao = yf.Ticker(acao)
    historico = yf_acao.history(start='2021-01-01',
                                end='2021-07-26')
    historico.Close.plot(label=acao)
    plt.title(acao)
    plt.legend()
#     plt.show()
```

```python
historico.to_excel('acao.xlsx')
```
