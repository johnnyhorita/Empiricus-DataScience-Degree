## 1 - Carregar os arquivos e cadastro, investimento1 e investimento2 e mostrar a quantidade de linhas de cada um, além das informações desses dataframes

```python
import pandas as pd
import datetime as dt
import numpy as np

df_cad = pd.read_csv('https://raw.githubusercontent.com/renatapink/DSdegree/master/Modulo2/Mini_projeto1/cadastro.csv')
df_invest1 = pd.read_csv('https://raw.githubusercontent.com/renatapink/DSdegree/master/Modulo2/Mini_projeto1/investimento_parte1.csv')
df_invest2 = pd.read_csv('https://raw.githubusercontent.com/renatapink/DSdegree/master/Modulo2/Mini_projeto1/investimento_parte2.csv')

print('Cadastros:')
df_cad.info()
print('\n---------------------------------------\n')
print('Investimeto1:')
df_invest1.info()
print('\n---------------------------------------\n')
print('Investimeto2:')
df_invest2.info()
```

## 2 - Juntar os dataframes de investimento1 e investimento2 em um só e ver a quantidade de linhas

```python
# Merge do tipo 'outer' porque desta forma já limpei os dados repetidos.
df_invest = df_invest1.merge(df_invest2, how='outer')

df_invest.duplicated().sum()

df_invest.info()
```

## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
df_cad_invest.info()
```

```python
colunas = np.array(df_cad_invest.columns)

def calc_na(col):
    if df_cad_invest[col].isnull().sum() > 0:
        na_in_col = df_cad_invest[col].isnull().sum()
        percent = na_in_col / df_cad_invest[col].size
        return format(percent, '.2f')
    else:
        pass

for i in colunas:
    na_percent = calc_na(i)
    print(f'{i}: {na_percent}')
```

```python
display(df_cad_invest[df_cad_invest.Nota1 == 0])
display(df_cad_invest[df_cad_invest.Nota2 == 0])

# Solução de alterar para 0 devido a inexistência de valores 0.
# Para operações futuras ignorar os valores iguais à 0 já que foi usado para preencher as faltas de valores.
df_cad_invest = df_cad_invest.fillna(0)
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50


```python
# Criando coluna 'Total' contendo a soma do valor investido em cada ativos.

df_cad_invest['Total'] = df_cad_invest['Ativo01'] + df_cad_invest['Ativo02'] + df_cad_invest['Ativo03'] + df_cad_invest['Ativo04'] + df_cad_invest['Ativo05'] + df_cad_invest['Ativo06'] + df_cad_invest['Ativo07']
```

```python
# Para não prejudicar a média final, transformo os valores 0 em nan novamente
# (antes de transformar os valores 'Nan' em 0 não existia nenhuma nota 0). 

df_cad_invest['Media'] = df_cad_invest[['Nota1', 'Nota2']].replace(0, np.nan).mean(axis=1, skipna=True)
```

```python
# Verificando o resultado e, tratando as médias Nan (Devido a valores Nan nas duas colunas) preenchendo com 0.

df_cad_invest.info()
df_cad_invest[(df_cad_invest['Nota2'] == 0) & (df_cad_invest['Nota1'] == 0)]
df_cad_invest.fillna(0, inplace=True)

df_cad_invest.info()
```

```python
# Separando os emails pelo '@' e transformando em um DataFrame e, adicionando a coluna contendo os dominios a nova coluna 'dominios'.

dominios = df_cad_invest['Email'].str.split('@', expand=True)
display(dominios[1])

df_cad_invest['Dominio'] = dominios[1]
df_cad_invest
```

```python
# Renomeando a coluna, removendo os '$' e já transformando os valores em float

df_cad_invest.rename(columns={'Salario': 'Renda'}, inplace=True)

df_cad_invest.Renda = df_cad_invest.Renda.str.replace('$', '').astype(float)

df_cad_invest
```

```python
# Verificando as possibilidades de sexo existentes e criando um dicionário com os valores para substituir.

display(df_cad_invest['Sexo'].unique())

dic_map = {
    'Fem': 'Feminino',
    'Masc': 'Masculino'
}
```

```python
# Substituindo os valores utilizando map e o 'dic-map' criado anteriormente.

df_cad_invest.Sexo = df_cad_invest.Sexo.map(dic_map)

df_cad_invest.Sexo
```

```python
# Arredondando as médias para o valor mais próximo e transformando em 'int'.

df_cad_invest.Media = df_cad_invest.Media.round(0).astype(int)
df_cad_invest.Media
```

```python
# Criando a coluna 'NPS' a partir de uma função lambda que realiza as verificações necesárias.

df_cad_invest['NPS'] = df_cad_invest.Media.apply(lambda x: 'Promotor' if x >= 80 else('Neutro' if ((x >= 50) and (x < 80)) else 'Defrator'))
```

```python
# Quantos investidores tem em cada estado (dica: use groupby)
df_cad_invest.groupby('Estado').size()
```

```python
# Qual é a soma do total investido por cada estado?
soma_estados = pd.DataFrame([(i, df_cad_invest[df_cad_invest.Estado == i].Total.sum()) for i in df_cad_invest.Estado.unique()])
soma_estados.set_index(soma_estados[0], inplace=True)
soma_estados.plot(kind='bar', xlabel='UF')
```

```python
# Como os investidores se distribuem em promotores, detratores e neutros?
df_cad_invest.groupby('NPS').size()
```

```python
# Qual é a média do total investido por promotores, detratores e neutros?
media_nps = pd.DataFrame([(i, df_cad_invest[df_cad_invest.NPS == i].Total.mean()) for i in df_cad_invest.NPS.unique()])
display(media_nps)
media_nps.set_index(media_nps[0]).plot(kind='bar', xlabel = 'NPS')
```
