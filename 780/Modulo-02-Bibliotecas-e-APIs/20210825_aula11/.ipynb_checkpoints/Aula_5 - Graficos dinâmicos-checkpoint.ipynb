## Line plots

Assim como o seaborn, o plotly oferece diversos datasets para experimentação!

```python
df_stocks = px.data.stocks()
```

```python
df_stocks
```

```python
px.line(df_stocks,
       x='date',
       y='GOOG')
```

```python
px.line(df_stocks,
       x='date',
       y=['GOOG', 'AAPL'])
```

```python
px.line(df_stocks,
       x='date',
       y=['GOOG', 'AAPL'],
       title='Apple vs Google')
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
                    x=df_stocks.date,
                    y=df_stocks.AAPL,
                    mode='lines',
                    name='Apple'
                ))

fig.add_trace(go.Scatter(
                    x=df_stocks.date,
                    y=df_stocks.AMZN,
                    mode='lines+markers',
                    name='Amazon'
                ))

fig.add_trace(go.Scatter(
                    x=df_stocks.date,
                    y=df_stocks.GOOG,
                    mode='lines',
                    name='Google',
                    line=dict(color='firebrick',
                              width=2,
                              dash='dashdot'  # dash, dot, dashdot
                          )
                      )
                  )
```
