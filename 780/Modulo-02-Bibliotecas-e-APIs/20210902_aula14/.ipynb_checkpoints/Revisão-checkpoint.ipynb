**Questão #1 - Matplotlib**

Crie um array x com 200 pontos, começando de -10 até 10. Depois calcule as seguintes funções sobre esse array:

y1 = sen(x) / x  
y2 = 1 / (1-e^(-x))

Por fim, plot essas duas funções, em função do x, ambas em uma mesma figura, porém, em dois axes distintos (um abaixo do outro).

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 200)
y1 = np.sin(x) / x
y2 = 1 / (1 - np.exp(-x))
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(5, 10), sharex=True)
ax1.plot(x, y1)
ax2.plot(x, y2)
```

**Questão #2 - Numpy**

Peça para o usuário uma lista com as comidas favoritas dele, depois transforme isso em um objeto numpy. Agora transformado, elimine todas as comidas que começam com uma vogal. Mostre o resultado.

Exemplo:

Entrada: [maçã, uva, morango, melancia, acelga]

Saída: [maçã, morango, melancia]

```python
import numpy as np
print('Por favor, insira as suas comidas favoritas...')
comida_favoritas = """maçã,uva,morango,melancia,acelga""".split(",")
arr_comida_favoritas = np.array(comida_favoritas)
vogais = ['a', 'e', 'i', 'o', 'u']

arr_letras_iniciais = np.char.ljust(arr_comida_favoritas, 1)
# np.char.rjust(comida_favoritas, 10)
# np.char.ljust(comida_favoritas, 10)
filtro = np.where(~np.in1d(arr_letras_iniciais, vogais))[0]
print(arr_comida_favoritas[filtro].tolist())
```

**Questão #10 - Numpy**

DESAFIO: Utilizando manipulação de arrays, ache os indexes do array cujo valor possua apenas 2 vogais no nome usando numpy.

```python
frutas = ['banana', 'morango', 'uva', 'pitaia', 'abacate', 'abacaxi', 'acerola', 'jenipapo', 'kiwi']
```

```python
frutas_arr = np.array(frutas)
vogais = ['a', 'e', 'i', 'o', 'u']
n_vogais = 2
contagem = sum([np.char.count(frutas_arr, vogal) for vogal in vogais])
indice_frutas = np.where(contagem==n_vogais)[0]
print(indice_frutas.tolist())
print(frutas_arr[indice_frutas])
```

```python
import pandas as pd
df = pd.DataFrame([np.char.count(frutas_arr, vogal) for vogal in vogais], index=vogais, columns=frutas)
soma = df.sum()
soma[soma==2].index.tolist()
```

```python
# Joins
```

```python
df1 = pd.DataFrame({
    'id': [1,2,3, 4],
    'fruta': ['banana', 'orange', 'apple', 'strawberry'],
})
df2 = pd.DataFrame({
    'fruta': ['banana', 'orange', 'apple', 'pineapple'],
    'calories': [200, 300, 400, 500]
})
```

```python
df1
```

```python
df2
```

```python
# Juntando pelas colunas
df1.merge(df2, on='fruta')  # inner join
```

```python
df1.merge(df2, on='fruta', how='left')  # left join
```

```python
df1.merge(df2, on='fruta', how='right')  # right join
```

```python
df1.merge(df2, on='fruta', how='outer')  # outer join
```

```python
df1.join(df2, lsuffix='_tabela1')  # Juntando tabelas pelo indice
```

```python
concat1 = pd.concat([df1, df2], axis=0)
concat1
```

```python
concat1.loc[0]
```

```python
concat2 = pd.concat([df1, df2], axis=1)
concat2
```
