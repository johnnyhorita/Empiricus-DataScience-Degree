Questão #1

Enunciado
Crie um array x com 200 pontos, começando de -10 até 10. Depois calcule as seguintes funções sobre esse array:

y1 = sen(x) / x
y2 = 1 / (1-e^(-x))

Por fim, plot essas duas funções, em função do x, ambas em uma mesma figura, porém, em dois axes distintos (um abaixo do outro).

```python
import matplotlib.pyplot as plt 
import math
import numpy as np

arr = np.linspace(-10, 10, 200)

y1 = []
y2 = []

for i in range(len(arr)):
    y1.append(math.sin(arr[i]) / arr[i])
    y2.append(1 / (1-10 ** (-arr[i])))
    i += 1

fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(arr, y1)
axes[1].plot(arr, y2)
plt.show() 
```

```python
plt.plot(arr, y1)
plt.plot(arr, y2)
plt.show()
```

```python
figure = plt.figure()
ax = figure.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(arr, y1, c='red', ls='--', lw=5, label='y1 = sen(x) / x')
ax.plot(arr, y2, c='#27a16c', marker='o', label='y2 = 1 / (1-e^(-x))')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Comparação')
ax.legend()
plt.show()
```

```python
plt.plot(y1, y2, color = 'red', marker = "o") 
plt.title("math.sin()") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.show() 
```

Enunciado
Carregue o arquivo avocado.csv e, em seguida, siga as instruções abaixo:

(a) Obtenha apenas os dados desse dataset que são referentes à região da Califórnia (California), no ano de 2017; e salva-os em um novo dataframe.
(b) A partir do dataframe obtido no item (a), obtenha apenas os dados referentes à abacates do tipo convencional; em seguida, salve em um novo dataframe.
(c) Utilizando o dataframe salvo no item (b), crie um gráfico de linha que apresente a variação do preço médio do abacate ao longo do ano de 2017 (datas) na região da Califórnia.
(d) Utilizando o dataframe salvo no item (a), crie um gráfico de pizza que apresente o percentual do volume total de abacates vendidos para cada tipo: convencional e orgânico.

Clique aqui para fazer o download do dataset.

```python
import matplotlib.pyplot as plt 
import pandas as pd

df = pd.read_csv ('https://s3-sa-east-1.amazonaws.com/lcpi/1f947ca4-bc5d-4c03-afff-53fe59a3bd97.csv', sep=',', encoding='utf8')

# (a) Obtenha apenas os dados desse dataset que são referentes à região da Califórnia (California), no ano de 2017;
# e salva-os em um novo dataframe.
#df["region"].unique()
filtro = ((df["region"]=="California") & (df["year"]==2017))
dfa = df[filtro]

# (b) A partir do dataframe obtido no item (a), obtenha apenas os dados referentes à abacates do tipo convencional; 
# em seguida, salve em um novo dataframe.
filtro = (dfa["type"]=="conventional")
dfb = dfa[filtro]

# (c) Utilizando o dataframe salvo no item (b), 
# crie um gráfico de linha que apresente a variação do preço médio do abacate ao longo do ano de 2017 (datas)
# na região da Califórnia.
dfb.plot(x='Date', y='AveragePrice', kind = 'line')

font1 = {'family':'serif','color':'blue','size':15}
font2 = {'family':'serif','color':'darkred','size':10}

plt.title("Média de Preços do Abacate Convencional", fontdict=font1, loc='center')
plt.xlabel("Data", fontdict=font2)
plt.ylabel("Preço médio", fontdict=font2)
plt.grid(axis='x', color = 'green', linestyle = '--', linewidth = 0.5)
plt.xticks(rotation=90)
plt.show()

# (d) Utilizando o dataframe salvo no item (a), 
# crie um gráfico de pizza que apresente o percentual do volume total de abacates vendidos para cada tipo: 
# convencional e orgânico.

dfpie = dfa.groupby(["type"]).agg(Total_Vol = ("Total Volume", "sum")).reset_index()
lbls = dfpie["type"].unique().tolist()

plt.figure(figsize=(16,8))

dfpie.plot(kind='pie', 
           y = 'Total_Vol',
           autopct='%1.1f%%',
           startangle=90,
           shadow=False,
           labels=lbls,
           legend = False,
           fontsize=14)

plt.title("Volume total de abacates vendidos", fontdict=font1, loc='center')
plt.ylabel("Percentual de vendas", fontdict=font2)
plt.show()
```

Carregue o arquivo houses_to_rent.csv. Em seguida:

(a) Crie um gráfico de dispersão para visualizar uma relação entre a área e o preço total do aluguel do imóvel.
(b) Visualize os outliers presentes na coluna da área utilizando o boxplot.
(c) Faça o mesmo que foi solicitado no item anterior, porém, para a coluna do preço total do aluguel.
(d) Elimine os outliers de ambas as colunas utilizando o método quantile. (e) Crie novamente um gráfico de dispersão dos dados da área e preço total do aluguel do imóvel; dessa vez, após a remoção dos outliers de ambas as colunas.

Clique aqui para fazer o download do dataset.

```python
import matplotlib.pyplot as plt 
import pandas as pd

df = pd.read_csv ('https://s3-sa-east-1.amazonaws.com/lcpi/bc1625bb-909a-4a15-8107-6e4047ec6cd4.csv', sep=',', encoding='utf8')

#renomeando os labels do dataframe
new_label = [item.replace(" ", "_") for item in [item.replace(" (R$)", "") for item in df.columns.tolist()]]
df.set_axis(new_label, axis=1, inplace=True)

print("a)")
plt.scatter(df.area, df.total)
plt.show()

print("b)")
fig = plt.figure(figsize =(10, 7))
plt.boxplot(df["area"])
plt.show()

print("c)")
fig = plt.figure(figsize =(10, 7))
plt.boxplot(df["total"])
plt.show()

print("d)")
print("Com outliers")
print(df.shape)
cols = ['area', 'total'] # one or more
Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
print("Retirando outliers")
print(df.shape)

print("e)")
plt.scatter(df.area, df.total)
plt.show()

```
