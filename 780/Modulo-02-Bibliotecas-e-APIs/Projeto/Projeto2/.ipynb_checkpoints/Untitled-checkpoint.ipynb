A característica do orçamento tem alguns valores impuros que fazem com que Pandas o atribua como um objecto genérico.
```python
data = [ dict(
        type = 'choropleth',
        locations = con_df['country'],
        locationmode = 'country names',
        z = con_df['num_movies'],
        text = con_df['country'],
        colorscale = [[0,'rgb(255, 255, 255)'],[1,'rgb(255, 0, 0)']],
        autocolorscale = False,
        reversescale = False,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            tickprefix = '',
            title = 'Production Countries'),
      ) ]

layout = dict(
    title = 'Production Countries for the MovieLens Movies (Apart from US)',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='d3-world-map' )
```

Datas de Lançamento de Filmes

As datas de lançamento podem muitas vezes desempenhar um papel muito importante na determinação do sucesso e das receitas geradas por um determinado filme. Nesta secção, tentaremos obter informações sobre as datas de lançamento em termos de anos, meses e dias da semana.

Já construímos a reportagem do ano na nossa etapa de discussão de dados preliminares. Vamos agora extrair o mês e o dia também para cada filme com uma data de lançamento.

```python
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
```

```python
def get_month(x):
    try:
        return month_order[int(str(x).split('-')[1]) - 1]
    except:
        return np.nan
```

```python
def get_day(x):
    try:
        year, month, day = (int(i) for i in x.split('-'))    
        answer = datetime.date(year, month, day).weekday()
        return day_order[answer]
    except:
        return np.nan
```

```python
df['day'] = df['release_date'].apply(get_day)
df['month'] = df['release_date'].apply(get_month)
```

Com estas características em mãos, vamos agora verificar os meses e dias mais populares e de maior sucesso.

```python
plt.figure(figsize=(12,6))
plt.title("Number of Movies released in a particular month.")
sns.countplot(x='month', data=df, order=month_order)
```

Parece que Janeiro é o mês mais popular quando se trata de lançamentos de filmes. Nos círculos de Hollywood, este é também conhecido como o mês do lixão, quando os filmes sub par são lançados pela dúzia.

Em que meses é que os filmes de "bockbuster" tendem a ser lançados? Para responder a esta pergunta, vamos considerar todos os filmes que tenham feito mais de 100 milhões de dólares e calcular a média bruta para cada mês.

```python
month_mean = pd.DataFrame(df[df['revenue'] > 1e8].groupby('month')['revenue'].mean())
month_mean['mon'] = month_mean.index
plt.figure(figsize=(12,6))
plt.title("Average Gross by the Month for Blockbuster Movies")
sns.barplot(x='mon', y='revenue', data=month_mean, order=month_order)
```

Vemos que os meses de Abril, Maio e Junho têm a maior média bruta entre os filmes de maior bilheteira. Isto pode ser atribuído ao facto de os filmes blockbuster serem geralmente lançados no Verão, quando as crianças estão fora da escola e os pais estão de férias e, portanto, é mais provável que o público gaste o seu rendimento disponível em entretenimento.

Será que alguns meses tendem a ser mais bem sucedidos do que outros? Vamos visualizar o boxplot entre o regresso e os meses.

```python
fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(15, 8))
sns.boxplot(x='month', y='return', data=df[df['return'].notnull()], palette="muted", ax =ax, order=month_order)
ax.set_ylim([0, 12])
```

Os meses de Junho e Julho tendem a produzir os maiores rendimentos medianos. Setembro é o mês com menos sucesso nas métricas acima mencionadas. Mais uma vez, o sucesso dos filmes de Junho e Julho pode ser atribuído ao facto de serem meses de Verão e períodos de férias. Setembro denota normalmente o início do semestre escolar/universitário e, portanto, uma ligeira redução no consumo de filmes.
