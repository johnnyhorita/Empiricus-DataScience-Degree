```python
Questão #23

Vale
1

Enunciado
Construa uma query que retorne quem foi o jogador que ficou por mais tempo como Top 1 do ranking e por quantos anos ele 
foi o Top 1.

SELECT position_player AS "Ranking", 
		COUNT(*) AS "Anos", 
		name_player AS "Jogador" 
FROM tb_chess 
WHERE position_player = 1
GROUP BY position_player, name_player
ORDER BY 1, 2 DESC
```
