```python
#Calculando todas as distâncias de todos os pontos de data para cada ponto de no_class

lista = []
def dist(lista1,lista2):
    for j in range(len(lista1)):
        for i in range(len(lista2)):
            d = ((lista2[i][2][0] - lista1[j][2][0])**2 + (lista2[i][2][1] - lista1[j][2][1])**2 + (lista2[i][2][2] - lista1[j][2][2])**2 + (lista2[i][2][3] - lista1[j][2][3])**2)**0.5
            lista.append(d)
                
    return  lista
```

```python
#Usando as informações sobre as distâncias para criar listas com as 120 distâncias calculadas entre um ponto
#em particular de no_class para cada ponto de data

def listasMenores(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
    return lst
        
listasDivididas = list(listasMenores(lista, 120))
print(listasDivididas)
```

```python
#Enumerar cada elemento de cada uma das listas criadas anteriormente

for i in range(len(listasDivididas)):
    for indice, distancia in enumerate(listasDivididas[i]):
        print(indice, distancia)
```

```python
#Transformando a os pares (indice,distância) em uma lista

lista_nova = []
for i in range(len(listasDivididas)):
    lista_nova.append(list(enumerate(listasDivididas[i])))
```

```python
#Criando um dicionário que tenha, como chave, o CPF, e como valores, as 5 menores distâncias 

results = {}
for cpf, distvec in zip(no_class, lista_nova):
    distvec_red = sorted(distvec, key=lambda x: x[1])[:5]
    results[cpf[0]] = distvec_red
```

```python
#Criando uma lista com os perfis correspondentes às posições da lista anterior

lista_perfis = []
for j in listasDivididas:
    for i in j:
        if i <= 39:
            lista_perfis.append('Conservador')
        if i >= 40 and i < 80:
            lista_perfis.append('Moderado')
        if i >=80 and i <120:
            lista_perfis.append('Agressivo')




lista_perfis
```

```python
lista_perfis_final = []

for j in lista_perfis2:
    contagem_C = lista_perfis2[lista_perfis2.index(j)].count('Conservador')
    contagem_M = lista_perfis2[lista_perfis2.index(j)].count('Moderado')
    contagem_A = lista_perfis2[lista_perfis2.index(j)].count('Agressivo')
    if contagem_C >= 3:
        lista_perfis_final.append('Conservador')
    if contagem_M >= 3:
        lista_perfis_final.append('Moderado')
    if contagem_A >= 3:
        lista_perfis_final.append('Agressivo')

        
lista_perfis_final
```
