# Matplotlib Aula 2

```python
# pyplot é um módulo do matplotlib que fornece funções para adicionar elementos à figura (linhas, imagens, texto...)
import matplotlib.pyplot as plt
import numpy as np

# para não precisar do plt.show(), ele invoca o gráfico
%matplotlib inline 
```

# Tipos de gráficos

```python
# Dados para usar nos gráricos
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
z = np.random.randn(1000)
t = [3, 6, 12, 20]

spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
```

```python
plt.plot(x, y)  # gráfico de linha
```

```python
plt.bar(x,y) # Gráfico de barras verticais
```

```python
plt.barh(x, y) # Gráfico de barras vertical
```

```python
plt.pie(x) # Gráfico de pizza
```

```python
plt.hist(z) # Histograma
```

```python
plt.scatter(x,y) # gráfico de disperção
```

```python
plt.boxplot(data) # Boxplot
```

```python
plt.fill_between(x, y) # gráfico de área
```

```python
 plt.stackplot(x,y,t) # gráfico empilhado
```

# Customizar/Personalizar o gráfico

```python
# São alguns exemplos
plt.figure(figsize=(16,10))

plt.plot(x, y, linestyle='-.', color='purple', linewidth=3, label='Nome da legenda')

plt.title('O nome do gráfico', color='red')

plt.xlabel('Nome do eixo x', color='b')
plt.ylabel('Nome do eixo y', color='green')

plt.xticks(rotation=45, color='blue')
plt.yticks(color='g', rotation=90)

plt.xlim(1, 4)
plt.ylim(2, 14)

plt.axhline(11, color='grey')
plt.axvline(1.8, color='yellow')

plt.grid()
plt.legend(loc=4)

plt.show()
```

Na estrutura do matplotlib temos dois conceitos importantes:
- Área de plotagem (eixos ou Axes) onde de fato os gráficos aparecem
- Figura (Figure) onde os eixos estão

# Subplots

```python
# Fazendo na mão - alternativa 1 - mais trabalhosa

# Definindo a área de plotagem
ax1 = plt.axes()
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
```

```python
# alternativa 2 - cria a área de plotagem e depois e vai adicionando os gráficos

# compartilhando o mesmo eixo x
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],  xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],  ylim=(-1.2, 1.2))

x2 = np.linspace(0,10)

ax1.plot(np.sin(x2))
ax2.plot(np.cos(x2)); # o ; tem a mesma função do plt.show() de mostrar a figura quando %matplotlib inline
```

como colocar os gráficos dentro desses eixos?

```python
# criando a estrutura

fig, axes = plt.subplots(1, 2)

#criar os gráficos em cada posição

axes[0].plot(x,y)
axes[1].hist(z);
```

```python
# criar os objetos de plotagem

fig, axes = plt.subplots(2, 5, figsize=(18,4))

# adicionar cada gráfico na posição

axes[0,0].plot(x, y)

axes[0,1].hist(z)

axes[0,2].barh(x,y)

axes[1,0].pie(x)

axes[1,1].bar(x,y)

axes[1,2].stackplot(x, y, t);
```

```python
# criar os objetos de plotagem

fig, axes = plt.subplots(2, 5, figsize=(18,6))

# adicionar cada gráfico na posição

axes[0,0].set_title('Gráfico de linhas')
axes[0,0].plot(x, y)

axes[0,1].set_title('Gráfico histograma')
axes[0,1].hist(z)

axes[0,2].set_title('Gráfico de barras horizontal')
axes[0,2].barh(x,y)

axes[1,0].set_title('Gráfico de pizza')
axes[1,0].pie(x)

axes[1,1].set_title('Gráfico de barras vertical')
axes[1,1].bar(x,y)

axes[1,2].set_title('Gráfico de área empilhado')
axes[1,2].stackplot(x, y, t);

axes[0,4].set_xlabel('eixo x')
axes[0,4].set_ylabel('eixo y')
```
