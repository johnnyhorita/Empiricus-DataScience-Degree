# Criar e manipular o dataframe

```python
df['variação'] = df['Close'] - df['Open']
```

```python
# outra forma de criar a coluna variacao

df.assign(variacao=lambda df: df['Close']- df['Open'])
```

```python
df['media'] = (df['High'] + df['Low'])/2
```

Obs: NaN --> Not a number : não é um número (nulo)

```python
# se a variação for negativa, a coluna var vai estar escrito negativo e o mesmo pensando para positivo

# criar coluna com condição

df.loc[df['variação'] < 0, 'var'] = 'negativa'
df.loc[df['variação'] > 0, 'var'] = 'positiva'
```

**MERGE** cruzando duas tabelas (dataframes) através de uma chave (coluna que tenha nas duas tabelas)

HOW do merge 

```python
# juntar os dados da df2 na df através de uma chave (Date)
df_novo = pd.merge(df, df2, on='Date', how='left') #on= chave how=tipo de junção, left, right, inner, outer
df_novo
```

```python
# Duvida, trazer apenas algumas colunas da tabela df2

# juntar os dados da df2 na df através de uma chave (Date)
pd.merge(df, df2[['Date', 'day_name']], on='Date', how='left') #on= chave how=tipo de junção, left, right, inner, outer
```

```python
df_novo
```

```python
# quero saber qual é o maior valor do Open das sextas feiras

df_novo[df_novo['day_name']=='Friday']['Open'].max()
```
