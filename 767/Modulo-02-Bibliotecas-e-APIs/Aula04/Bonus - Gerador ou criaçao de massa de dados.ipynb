```python
import pandas as pd
import numpy as np
import random
```

# Como criar ou gerar dados?

```python
# 10k investidor

id_investidor = [str(i).zfill(7) for i in range(1,100001)]

lst_perfil = [random.choice(['Moderado','Conservador', 'Agressivo']) for i in range(100000)]

# usando random.gauss para  valores de média 1000 e desvio padrão 300. round(n,2) arredonda para 2 casas decimais
# fazer isso 10000 vezes
carteira01 = [round(random.gauss(1000, 300),2) for i in range(100000)] 
carteira02 = [round(random.gauss(5000, 500),2) for i in range(100000)]
carteira03 = [round(random.gauss(3000, 500),2) for i in range(100000)]
carteira04 = [round(random.gauss(10000, 10000),2) for i in range(100000)]
carteira05 = [round(random.gauss(5000, 1000),2) for i in range(100000)]
carteira06 = [round(random.gauss(2000, 400),2) for i in range(100000)]
```

```python
# !pip install br-nome-gen

from br_nome_gen import pessoa_random

# p = pessoa_random()
# print(f"{'Homem' if p.masc else 'Mulher'}: {p.nome}")

lst_nome = []
lst_masc = []
for i in range(0, 100000):
    p = pessoa_random()
    lst_nome.append(p.nome)
    lst_masc.append(p.masc)
```

```python
# !pip install Faker
from faker import Faker
fake = Faker()

lst_date = []
lst_latlong = []
lst_job = []
lst_email = []
lst_pricetag = []
for _ in range(100000):
    lst_date.append(fake.date())
    lst_latlong.append(fake.latlng())
    lst_job.append(fake.job())
    lst_email.append(fake.ascii_free_email())
    lst_pricetag.append(fake.pricetag())
```

```python
# lembrando do numpy array 10000 x 2 
np.random.random_sample((100000, 2)) * 100
```

```python
df = pd.concat([
    pd.DataFrame.from_dict({
        'id': id_investidor,
        'Nome': lst_nome,
        'Sexo': lst_masc,
        'Nasc': lst_date,
        'Email': lst_email,
        'Lat_long': lst_latlong,
        'Job': lst_job,
        'Pricetag': lst_pricetag,
        'Perfil': lst_perfil,
        'carteira01': carteira01,
        'carteira02': carteira02,
        'carteira03': carteira03,
        'carteira04': carteira04,
        'carteira05': carteira05,
        'carteira06': carteira06
    }),
    pd.DataFrame(np.random.random_sample((10000, 2)) * 100,
                 columns=['nota1', 'nota2'])
],
               axis=1)
```
