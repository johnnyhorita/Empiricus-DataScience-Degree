## Biblioteca e importação dos dados

Dentro do seaborn já tem alguns datasets embutidos e basta acessá-los com o método load_dataset

https://github.com/mwaskom/seaborn-data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
# Carregar dataset dentro da biblioteca

df = sns.load_dataset('penguins')
df
```

- species: espécies de pinguins (Chinstrap, Adélie, or Gentoo)
- bill_length_mm: comprimento do bico (mm)
- bill_depth_mm: profundidade do bico (mm)
- flipper_length_mm: comprimento da nadadeira (mm)
- body_mass_g: massa corporal (g)
- island: nome da ilha (Dream, Torgersen, or Biscoe) no arquipélogo Palmer (Antarctica)
- sex: sexo do pinguim

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include='O')
```

```python
# eliminar os dados faltantes
df = df.dropna()
```

obs: dropna() elimina os dados faltantes

```python
# Gráfico de barras de frequencia
sns.catplot(data=df, x='sex', kind='count')
plt.title('Quantidade de Piguins do sexo masculino e feminino')
plt.show()
```

```python
# No matplotlib
df['sex'].value_counts().plot(kind='bar')
plt.title('Quantidade de Piguins do sexo masculino e feminino')
plt.show()
```

distribuição dos dados

```python
# Histograma - mostra a distribuição dos dados
sns.displot(df['body_mass_g'], kind='hist')
```

```python
sns.displot(df['body_mass_g'], kind='kde')
```

```python
sns.displot(df['body_mass_g'], kind='hist', kde=True)
```

```python
# distribuição acumulada dos dados
sns.displot(df['body_mass_g'], kind='ecdf')
```

Múltiplos gráficos

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))

sns.boxplot(x='species', y='bill_length_mm', data=df, ax=ax[0][0])
sns.boxplot(x='species', y='bill_depth_mm', data=df, ax=ax[0][1])
sns.boxplot(x='species', y='flipper_length_mm', data=df, ax=ax[1][0])
sns.boxplot(x='species', y='body_mass_g', data=df, ax=ax[1][1])
```

```python
# exemplo de pinguins outlier
df[(df['species']=='Chinstrap')&(df['body_mass_g']<3000)]
```
