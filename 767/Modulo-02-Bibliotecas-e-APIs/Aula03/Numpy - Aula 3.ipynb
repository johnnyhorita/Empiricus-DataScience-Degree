# Combinando array

`np.concatenate((a, d), axis=0)`  # Concatenate arrays  
`np.vstack((a, b))`  #Stack arrays vertically(rowwise)  
`np.r_[e, f]`  ##Stack arrays vertically(rowwise)  
`np.hstack((e, f))`  #Stack arrays horizontally(columnwise)  
`np.column_stack((a, d))`  #Create stacked columnwise arrays  
`np.c_[a, d]`  #Create stacked columnwise arrays  

```python
display(a, b, d)
```

```python
np.concatenate((a, d), axis=0)
```
