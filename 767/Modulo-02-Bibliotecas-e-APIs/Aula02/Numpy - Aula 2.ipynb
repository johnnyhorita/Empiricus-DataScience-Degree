2D

- 0 Ã© linha e 1 coluna para array de 2 D
- AtenÃ§Ã£o: vocÃª percorre o eixo, Ã© atravÃ©s do eixo
- axis Ã© o eixo de agragaÃ§Ã£o

```python
b = np.array([[1, -2, 3], [4, 5, -6]])
b
```

```python
np.sum(b) #numpy.sum(a, axis=None, dtype=None, out=None, keepdims=<no value>, initial=<no value>, where=<no value>)
# Sum of array elements over a given axis.
```

```python
np.sum(b, axis=0) # somando por meio do eixo 0, a linha
```

```python
np.sum(b, axis=1) # somando  1+(-2)+3 = 2 (que Ã© a linha 0) e 4+5+(-6)=3 (que Ã© a linha 1) estou agregando(somando) pelo eixo1 
# somando todos os elementos da linha 0 e linha 1
```

```python
np.min(b, axis=1)
```

```python
# funciona min, max, mean, cumsum, std, sort para 2D
b.sort(axis=1) # ordenar o array b pelo eixo 1
b
```
