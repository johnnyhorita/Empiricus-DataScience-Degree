```python
# Questão 1
# Crie uma classe Bola cujos atributos são cor e raio. 
# Crie um método que imprime a cor da bola. 
# Crie um método para calcular a área dessa bola. 
# Crie um método para calcular o volume da bola. 
# Crie um objeto dessa classe e calcule a área e o volume, imprimindo ambos em seguida.
#
# Obs.:
# Área da esfera = 4*3.14*r*r;
# Volume da esfera = 4*3.14*r*r*r/3

class Bola:
    # criando o método construtor
    def __init__(self, cor, raio):
        # cria variável (atributo) dentro do objeto
        self.cor = cor
        self.raio = raio
        self.pi = 3.14

    def __repr__(self):
        return 'Bola (cor=%s, raio=%s)' % (self.cor, self.raio)
    
    '''
    Metodo imprimir
    Retorna o print 'Cor da bola'
    '''
    def imprimir(self):
        print(f'Cor da bola {self.cor}')

    '''
    Metodo calcularArea
    Retorna o calculo da área da bola
    '''
    def calcularArea(self):
        a =  4 * self.pi * (self.raio ** 2)
        return a

    '''
    Metodo calcularVolume
    Retorna o calculo o volume da bola
    '''
    def calcularVolume(self):
        v =  (4 * self.raio ** 3) * self.pi /3
        return v

bola = Bola('Vermelha', 5)

bola.imprimir()
print(f'Area: {bola.calcularArea()}, Volume: {bola.calcularVolume()}')
bola
```

```python
# Questão 4
# Crie uma classe Televisor cujos atributos são:
# a. fabricante;
# b. modelo;
# c. canal atual;
# d. lista de canais; e
# e. volume.
# Faça métodos para aumentar/diminuir volume, 
# trocar o canal e 
# sintonizar um novo canal, que adiciona um novo canal à lista de canais (somente se esse canal não estiver nessa lista). 
# No atributo lista de canais, devem estar armazenados todos os canais já sintonizados dessa TV.
# Obs.: O volume não pode ser menor que zero e maior que cem; só se pode trocar para um canal que já esteja na lista de canais.

class Televisor:
    # criando o método construtor
    def __init__(self, fabricante, modelo):
        # cria variável (atributo) dentro do objeto
        self.fabricante = fabricante
        self.modelo = modelo
        self.canalatual = 0
        self.volume = 0
        self.canallista=[]
        self.selvolume = 5

    def __repr__(self):
        return 'Televisor (fabricante=%s, modelo=%s, canalatual=%s, volume=%s, canallista=%s)' % (self.fabricante, self.modelo, self.canalatual, self.volume, self.canallista)
        
    '''
    Metodo controlarVolume
    Retorna o calculo
    '''
    def controlarVolume(self, botao, repeticao=1):
        volume = self.volume
        if botao == '+':
            volume += repeticao
        elif botao == '-':
            volume -= repeticao
                    
        if volume >= 100:
            self.volume = 100
        elif volume <= 0:
            self.volume = 0
        else:
            self.volume = volume
        return self.volume
    
    def controlarCanal(self, canal):
        if canal in self.canallista:
            self.canalatual = canal
        else:
            print(f'O canal {canal} não foi sintonizado!')
            self.canalatual = 0
        return 
    
    def sintonizarCanal(self, canal):
        if canal not in self.canallista:
            print(f'O canal {canal} esta sintonizado!')
            self.canallista.append(canal)
        else:
            print(f'O canal {canal} já está sintonizado!')
        return self.canallista

televisor = Televisor('Samsung', 'XPTO') 
televisor.controlarVolume('+')
televisor.sintonizarCanal(5)
televisor.controlarVolume('+', 10)
televisor.sintonizarCanal(12)
televisor.sintonizarCanal(4)
televisor.controlarCanal(5)
televisor
```
