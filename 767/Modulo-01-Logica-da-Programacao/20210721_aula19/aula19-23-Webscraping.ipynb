```python
# criar uma função para buscar info em cada página:

def pagina_ativos(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    acao = []
    codigo = []
    for nome, ticker in zip( soup.find_all(class_='String Column1'), soup.find_all(class_='String Column2 ColumnLast')):
        acao.append(nome.get_text())
        codigo.append(ticker.get_text())
    
    return acao, codigo
    
```

```python
paginas = []
for href in soup.find_all('div', {'class': 'link_list atoz_list'})[0].find_all('a', href=True):
    link = 'https://br.advfn.com' + href['href']
    paginas.append(link)
```

```python
lst_page_acao = []
lst_page_codigo = []

for url in paginas:
    print(url)
    acao, codigo = pagina_ativos(url)
    lst_page_acao.extend(acao)
    lst_page_codigo.extend(codigo)
    
    print(len(lst_page_acao), len(lst_page_codigo))
```
