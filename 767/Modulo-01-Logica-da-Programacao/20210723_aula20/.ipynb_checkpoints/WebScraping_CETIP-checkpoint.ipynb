```python
def validaData(date_text):
    """
    Função para validar se a string recebida é do tipo data

    Argumentos:
        date_text: data no formato texto dd/mm/aaaa

    Retorno:
        Booleano True ou False

    Exemplo: 
        date_text = '02/01/2021'
        True
        validaData('02/01/2021')
    """
    try:
        datetime.datetime.strptime(date_text, '%d/%m/%Y')
    except ValueError:
        return False
    return True


def listaCabecalho(nmClass):
    """
    Função para definir o cabecalho de atributos da página

    Argumentos:
        nmClass define o nome da tag class do html
        nmClass: string

    Retorno:
        Lista com os atributos do cabecalho

    Exemplo: 
        nmClass = 'ConsultaCabecalho_02'
        listaCabecalho('ConsultaCabecalho_02')
    """
    listaCabecalho = list()
    for e in soup.find_all('td', {'class': nmClass}):
        listaCabecalho.append(e.get_text())
    #A pagina tras o cabecalho em 2 linhas 
    #Os comandos abaixo deixam a lista com a mesma ordem do cabecalho da página
    listaCabecalho.append(listaCabecalho[2]) # Repete o item 'Taxa SELIC' na última posição da lista
    del listaCabecalho[2] # Remove o item 2 'Taxa SELIC'
    del listaCabecalho[1] # Remove o item 1 'Taxas'
    return listaCabecalho

    
def listaSerie(nmClass):
    """
    Função para listar as séries encontradas na página

    Argumentos:
        nmClass define o nome da tag class do html
        nmClass: string

    Retorno:
        Lista com as séries da página

    Exemplo: 
        nmClass = 'ConsultaDados_R_01'
        listaSerie('ConsultaDados_R_01')
    """
    lista = list()
    listaSerie = list()
    for e in soup.find_all('td', {'class': nmClass}):
        lista.append(e.get_text())    
        #print(e.get_text())
    lista = [e.replace('\xa0\xa0', '') for e in lista]
    
    listaSerie = list()
    for e in lista:
        if validaData(e):
            linha = list()
            listaSerie.append(linha)
            linha.append(e)
        else:
            linha.append(str(e))
    return listaSerie


def dataframeSerie(listaSerie_1, listaSerie_2, cabecalho):
    """
    Função para agrupar as séries das listas das páginas na página, 
    pois são divididas entre datas pares e impares

    Argumentos:
        listaSerie_1 lista de séries capturadas pela função listaSerie (par ou impar)
        listaSerie_2 lista de séries capturadas pela função listaSerie (par ou impar)
        cabecalho lista do cabecalho capturadas pela função listaCabecalho

        listaSerie_1: list
        listaSerie_2: list
        cabecalho: list
        
    Retorno:
        Dataframe consolidado com as listaSerie_1 e listaSerie_2 com o cabecalho 

    Exemplo: 
        cabecalho = listaCabecalho('ConsultaCabecalho_02')
        listaSerie_1 = listaSerie('ConsultaDados_R_01')
        listaSerie_2 = listaSerie('ConsultaDados_R_02')
        
        dataframeSerie(listaSerie_1, listaSerie_2, cabecalho)
    """
    dataframeSerie = pd.DataFrame()
    df1 = pd.DataFrame(listaSerie_1, columns=cabecalho)
    df2 = pd.DataFrame(listaSerie_2, columns=cabecalho)
    dataframeSerie = pd.concat([df1.reset_index(drop=True), df2.reset_index(drop=True)], ignore_index=True)
    return dataframeSerie
```

```python
# instancia o objeto serie com a classe TestSerieHistoricaDI
serie = TestSerieHistoricaDI()

# Inicializa o browser 
#serie.bEdge() # Inicializa o browser Edge
serie.bChrome() # Inicializa o browser Chrome
#serie.bBrave() # Inicializa o browser Brave

# preenche os atributos do site de Série Histórica DI - CETIP 
# de: dd, mm, aaaa - até: dd, mm, aaaa
serie.serieHistoricaDI("01","06","2018","30","06","2020")

# define a página de origem para captura de informações
url = serie.driver.page_source
soup = BeautifulSoup(url, 'html.parser') # instancia objeto BeautifulSoup
```

```python
# labels da página
labelDI = soup.find_all('h3')[0].get_text() # Header 
labelTxPre = soup.find_all('h4')[0].get_text()
labelDtFiltro = soup.find_all('label')[0].get_text().replace('\t', '').replace('\xa0', '').replace('\n    ',' ').replace('Até:', ' Até:')

print(labelDI)
print(labelTxPre)
print(labelDtFiltro)

dfSerie = pd.DataFrame()
i = 1
print()
print(f'Página {i}')

importaSerie = True
while importaSerie:
    url = serie.driver.page_source
    soup = BeautifulSoup(url, 'html.parser')
    df = pd.DataFrame()
    
    cabecalho = listaCabecalho('ConsultaCabecalho_02')
    listaSerie_1 = listaSerie('ConsultaDados_R_01')
    listaSerie_2 = listaSerie('ConsultaDados_R_02')
    df = dataframeSerie(listaSerie_1, listaSerie_2, cabecalho)
    
    if serie.NextPage():
        i += 1
        print(f'Página {i}')
        dfSerie = dfSerie.append(df, ignore_index=True)
    else:
        print('\nWebScraping finalizado.')        
        dfSerie = dfSerie.append(df, ignore_index=True)
        importaSerie = False        

dfSerie['Dt'] = pd.to_datetime(dfSerie['Data'], dayfirst=True)
dfSerie = dfSerie.sort_values(by=['Dt'], ascending=True)
del dfSerie["Dt"]

# Salvar o dataframe
dt = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
nmArquivo = dt + '_TaxasDIPREOver.csv'

dfSerie.to_csv(nmArquivo, index=False, encoding='utf-8-sig', sep=';')
```
