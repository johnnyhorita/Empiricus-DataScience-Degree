## Amostragem

- Amostra é um subgrupo da população


- Tipos de amostragem:
    - Probabilísticas: Amostragem Aleatória Simples, Amostra Sistemática, Amostra Estratificada, Amostra por conglomerado 
    - Não probabilísticas: Amostra por conveniência, Amostra por julgamento, Amostra por cotas, Amostra bola de neve, Amostra despoporcional

**Amostragem Aleatória Simples**:<br>
Cada indivíduo da população tem a mesma chance de ser escolhido. A seleção de elementos é feita em forma de sorteio, dessa forma, não há critério ou filtro no processo de amostragem.    


**Amostragem Sistemática**: <br>
Há um fator matemático de seleção dos indivíduos.

**Amostragem Estratificada**: <br>
Os grupos da população são segmentados com características próprias e depois são selecionados indivíduos dentro desses grupos proporcionalmente.

**Amostragem por conglomerados**: <br>
A população é divida em grupos (subpopulação) e a amostragem é deita sobre os conglomerados (cluster) e não mais sobre os indivíduos da população
